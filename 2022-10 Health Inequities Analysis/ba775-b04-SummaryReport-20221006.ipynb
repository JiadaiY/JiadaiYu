### 2.3  Neighborhoods with different health place index(hpi)

```python
%%bigquery
WITH all_nopris AS(
    SELECT 
        county,
        date,
        metric_value AS positive_rate_all_nopris
    FROM `ba775-team-project-04.ba775_team04_covid_19_equity_metrics_v1kpo4.covid-19-health-equity-metric-pos-30-day-by-cnt`
    WHERE metric = 'county_positivity_all_nopris'
        AND metric_value IS NOT NULL),

 low_hpi AS(
    SELECT 
        county,
        date,
        metric_value AS positive_rate_low_hpi
    FROM `ba775-team-project-04.ba775_team04_covid_19_equity_metrics_v1kpo4.covid-19-health-equity-metric-pos-30-day-by-cnt`
    WHERE metric = 'county_positivity_low_hpi'
         AND metric_value IS NOT NULL)

SELECT 
    all_nopris.county,
    all_nopris.date,
    positive_rate_all_nopris,
    positive_rate_low_hpi,
    (positive_rate_low_hpi - positive_rate_all_nopris) as positive_rate_diff
FROM all_nopris
INNER JOIN low_hpi
ON all_nopris.county = low_hpi.county
    AND all_nopris.date = low_hpi.date
```

Neighborhoods with low healthy places index (HPI) tends to have higher positive rate than the county average, indicating more severe Covid epidemic and inadequate testing. Further, the higher the positivity rate, the more significant the differences in positivity rates across regions, while counties with overall lower positivity rates had smaller differences in positivity rates across regions.
The Covid-19 epidemic reveals health inequalities across neighborhoods, and the worse the epidemic, the more significant the inequalities

### 2.4  Race and Ethnicity

In this part,we are going to examine  the case rate and death rate based on different ethnical groups.
This dataset includes COVID-19 related data of California State from 07/04/2020 to 09/10/2022.
The columns of this dataset is explained as follows:
Date: exact date of recording, this dataset takes weekly recorded cases.

Location:The location of the record taken.In our particular dataset, the location is listed as CA.

Location level:The geographic level of the location.In our dataset,the location level is state.

Demographic: The demographic characteristic of all records taken.In our dataset,it is race-ethnity.

Demographic category: The exact division of different ethnicities.We have seven categories: 
white,Latino,Multi-race,Asian-American,American-Indian,African-American and Native Hawaiian.

Metric category:The metric we use when we do the analysis,including number of tests taken, 
number of confirmed cases and number of deaths.

Metric: The metric of the value in our dataset. Cases per week simply indicates number of cases per 100K 
and is not accumulated.

Value:The value based on metrics.

Question:
Is there a pattern that shows differences in total cases confirmed in 2021 between ethnicities? What may cause this difference ?

```python
%%bigquery
select sum(value) as total_cases_per100K,DEMOG_CAT
from `ba775-team-project-04.ba775_team04_covid_19_equity_metrics_v1kpo4.covid-19-race-ethnicity-timeseries`

where METRIC_CAT = 'cases'
and EXTRACT(year from DATE) = 2021

group by DEMOG_CAT
order by total_cases_per100K desc
```

The above results show significant difference in total cases within different ethnicity groups. The relative high
number of confirmed cases in Native Hawaiian may due to the fact that this ethnicity group lives in a certain area
and is easily get infected if the virus has spread to their living habitat.
The relative low number of confirmed cases in Asian American may due to the fact that they tend to be more cautious facing this virus and did more protective measures compared to other ethnicity groups.

Question:
Is there a pattern that shows differences in total death number in 2021 between ethnicities? 
What may cause this difference ?

```python
%%bigquery
select sum(value) as total_deaths_per100K,DEMOG_CAT
from `ba775-team-project-04.ba775_team04_covid_19_equity_metrics_v1kpo4.covid-19-race-ethnicity-timeseries`

where METRIC_CAT = 'deaths'
and EXTRACT(year from DATE) = 2021

group by DEMOG_CAT
order by total_deaths_per100K desc
```
